```python
import nltk
from nltk.corpus import names
import random
```

```python
nltk.download('names')
```

```python
male_names=[(name,'male') for name in names.words ('male.txt')]
female_names=[(name,'female') for name in names.words('female.txt')]
```

```python
all_names=male_names+female_names
random.shuffle(all_names)
```

```python
def gender_features(word):
    return {
        'last_letter': word[-1],  
        'first_letter': word[0],    
        'last_two': word[-2:],       
        'length': len(word)          
    }
```

```python
featuresets = [(gender_features(name), gender) for (name, gender) in all_names]
```

```python
train_set=featuresets[5000:]
test_set=featuresets[:2000]
```

```python
from nltk.classify import NaiveBayesClassifier

classifier = NaiveBayesClassifier.train(train_set)
```

```python
from nltk.classify.util import accuracy
print("Accuracy:",accuracy(classifier, test_set))
```

```python
print(classifier.classify(gender_features("Shreya")))
print(classifier.classify(gender_features("Arjun")))
print(classifier.classify(gender_features("John")))
print(classifier.classify(gender_features("Priya")))
```

```python
classifier.show_most_informative_features(10)
```
